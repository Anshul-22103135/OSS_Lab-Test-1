```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sci
```

```python
a = np.array([100, 120, 85, 90, 110, 95])
```

```python
b = np.array([80, 75, 60, 95, 85, 90])
c = np.array([150, 140, 135, 160, 155, 170])
```

```python
sum_a = sum(a)
print("Total Rainfall in City A:",sum_a)
sum_b = sum(b)
print("Total Rainfall in City B:",sum_b)
sum_c = sum(c)
print("Total Rainfall in City C:",sum_c)
```

```python
avg_a = sum_a/a.size
print("Monthly Average Rainfall of City A :",avg_a)
avg_b = sum_b/b.size
print("Monthly Average Rainfall of City B :",avg_b)
avg_c = sum_c/c.size
print("Monthly Average Rainfall of City C :",avg_c)
```

```python
def avg_all(a,b,c,i):
    avg = a[i] + b[i] + c[i]
    return avg/3
```

```python
i = int(input("Enter the number of month:"))
print("The Average of Given month is",avg_all(a,b,c,i))
```

```python
a1 = np.empty(6)
```

```python
def avg_p(a,b,c):
    for i in range(0,6):
        new_avg = a[i] + b[i] + c[i]
        a1[i] = new_avg/6
    
print("The Monthly averge rainfall for each month is:",avg_p(a,b,c))
for i in range (0,6):
    print("The Average Rainfall for Month ",i+1,"is",a1[i])
    
```

```python
y = [1, 2, 3, 4, 5, 6]
plt.plot(y,a,label="City A")
plt.plot(y,b,label="City B")
plt.plot(y,c,label="City C")
plt.xlabel("Number of Month")
plt.ylabel("Amount of Rainfall(in mm)")
plt.title("Month-wise Trend Of Rainfall on 6 cities")
plt.legend()
plt.show()
```

```python
def sortt(g):
    for i in range(0,6):
        j = i+1
        for j in range(0,6):
            if g[i] < g[j]:
                t = g[i]
                g[i] = g[j]
                g[j] = t
    return g
```

```python
new_a = sortt(a)
new_b = sortt(b)
new_c = sortt(c)
r_a = new_a[-1] - new_a[0]
r_b = new_b[-1] - new_b[0]
r_c = new_c[-1] - new_c[0]
print("Range of Rainfall for City A:",r_a)
print("Range of Rainfall for City B:",r_b)
print("Range of Rainfall for City C:",r_c)
```
